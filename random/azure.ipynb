```python
from diagrams import Diagram, Cluster
from diagrams.azure.general import Managementgroups as MGS
from diagrams.azure.general import Subscriptions as Subs
from IPython.display import Image

with Diagram("PINS Azure Architecture", show=False, direction="TB"):
    # with Cluster("RTL"):
    #     subs_rtl = [Subs("Dev"),
    #                 Subs("Test")]
        
    # with Cluster("Live"):
    #     subs_live = [Subs("Prod")]

    with Cluster("Groups"):
        mgmtgrps = [MGS("PINS RTL"),
                    MGS("PINS Live")]
        
    mgroups = [
        "PINS RTL",
        "PINS Live"
    ]

    for i in mgroups:
        with Cluster(f"Subs - {i}"):
            if i == "PINS RTL":
                subs_rtl = [Subs("Dev"),
                            Subs("Test")]
            elif i == "PINS Live":
                subs_live = [Subs("Prod")]

    MGS("PINS Root") >> mgmtgrps
    MGS("PINS RTL") >> subs_rtl
    MGS("PINS Live") >> subs_live
    # MGS("Tenant root") >> MGS("PINS root") >> MGS("PINS RTL"), MGS("PINS Live")
    # mgmtgrps >> subs_rtl
    # mgmtgrps >> subs_live
```
