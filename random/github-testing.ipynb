```python
from github import Github
import pandas as pd
import pprint
print("Imports complete")
```

```python
# Specify the organization or user
organization_name = 'Planning-Inspectorate'

# Replace with your GitHub access token
github_token = ''
```

```python
# Initialize the GitHub API client
g = Github(github_token)
```

```python
# Function to perform the compliance checks
def perform_compliance_checks():

    compliance = []

    # Get the organization or user
    org_or_user = g.get_organization(organization_name)

    # Loop through repositories in the organization or user
    for repo in org_or_user.get_repos():
        # if repo.name == 'ODW-Service':
        repo_name = repo.name

        # Check if CODEOWNERS file exists
        try:
            codeowners_file = repo.get_contents('CODEOWNERS')
            codeowners_exists = True
        except Exception as e:
            codeowners_exists = False

        # Get default branch
        default_branch = repo.default_branch

        # Check branch protection
        try:
            branch = repo.get_branch(default_branch)
            branch_protection = branch.protected
            if branch_protection == True:
                branch_protected = True
            else:
                branch_protected = False
        except Exception as e:
            print(f'{repo_name} -> {e}')

        # Check license
        try:
            repo_license = repo.license.name
        except Exception as e:
            repo_license = 'No license'
            print(f'{repo_name} -> {e}')

        # Check releases
        try:
            no_of_releases = len(list(repo.get_releases()))
        except Exception as e:
            print(f'{repo_name} -> {e}')

        compliance.append((repo_name, codeowners_exists, default_branch, branch_protected, repo_license, no_of_releases))

    return compliance
```

```python
df_cols = ['repo_name', 'codeowners_exists', 'default_branch', 'branch_protected', 'license', 'no_of_releases']
df = pd.DataFrame.from_records(perform_compliance_checks(), columns=df_cols)
df.sort_values(by=['repo_name'], ascending=True)
```
